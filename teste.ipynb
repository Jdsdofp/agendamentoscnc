```python
import xlrd
from collections import OrderedDict
import json

wb = xlrd.open_workbook("/bd2022.xlsx")

ws = wb.sheet_by_name("2022_bd")

data_list = []

sh = wb.sheet_by_index(0)

for rownum in range(1, sh.nrows):
  data = OrderedDict()

  row_values = sh.row_values(rownum)
  #print(row_values)

  data_list.append(data)

  data['Empresa'] = row_values[0]
  data['Fornecedor'] = row_values[1]
  data['Servico'] = row_values[2]
  data['Tipo'] = row_values[3]
  data['Venc'] = row_values[4]
  data['Janeiro'] = row_values[5]
  data['Fevereiro'] = row_values[6]
  data['Maeco'] = row_values[7]
  data['Abril'] = row_values[8]
  data['Maio'] = row_values[9]
  data['Junho'] = row_values[10]
  data['Julho'] = row_values[11]
  data['Agosto'] = row_values[12]
  data['Setembro'] = row_values[13]
  data['Outubro'] = row_values[14]
  data['Novembro'] = row_values[15]
  data['Dezembro'] = row_values[16]


  data_list.append(data)


  j = json.dumps(data_list, indent = 4)
  print(j)

with open(r'data', 'w') as f:
    f.write(j)
    print(j)
```
